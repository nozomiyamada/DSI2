# import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (10, 8)
import seaborn as sns
from pandas_profiling import ProfileReport
```

# load cleaned data

```python
act = pd.read_csv('data/act.csv') # 51 states * 3 years
act
```

# visualization

```python
act.groupby('Year')[['Participation','Composite']].corr()
```

```python
sns.scatterplot(
    y='Participation',
    x='Composite',
    hue='Year',
    palette='dark',
    data=act)
```

```python
plt.figure(figsize=(20, 15))
sns.barplot(
    y='State',
    x='Participation',
    hue='Year',
    order=act.groupby('State')['Participation'].mean().sort_values(ascending=False).index,
    data=act)
```
